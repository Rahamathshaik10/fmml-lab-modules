```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
# Initialize starting point
new_x = 3  # Initial guess
learning_rate = 0.1  # Learning rate

# Gradient descent
for n in range(500):
    previous_x = new_x

    # Calculate gradient at the current point
    gradient = 2 * (previous_x - 5)  # Derivative of f(x) = (x - 5)^2

    # Update x using gradient descent formula: x = x - learning_rate * gradient
    new_x = previous_x - learning_rate * gradient

    # Stopping criterion - check for convergence
    if abs(new_x - previous_x) < 1e-6:
        print(f'Converged at iteration {n} with x = {new_x}')
        break
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its derivative
def f(x):
    return x**2 + x + 1

def df(x):
    return 2*x + 1

# Gradient Descent
def gradient_descent(starting_point, learning_rate, iterations):
    x_list = [starting_point]
    slope_list = [df(starting_point)]
    for i in range(iterations):
        new_x = x_list[-1] - learning_rate * df(x_list[-1])
        x_list.append(new_x)
        slope_list.append(df(new_x))
    return x_list, slope_list

# Parameters for gradient descent
starting_point = -2
learning_rate = 0.3
iterations = 20

# Run gradient descent
x_list, slope_list = gradient_descent(starting_point, learning_rate, iterations)

# Create x values for the plots
x_1 = np.linspace(-3, 3, 1000)

# Superimpose the gradient descent calculations on the plot
plt.figure(figsize=[20, 5])

# 1 Chart: Cost function
plt.subplot(1, 3, 1)
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)
values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 3, 2)
plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# 3 Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)
plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the derivative function
def dg(x):
    return 2 * x  # Derivative of f(x) = x^2 is f'(x) = 2x

# Gradient descent function
def gradient_descent(derivative_func, initial_guess, multiplier=0.1, precision=0.0001, max_iter=1000):
    current_x = initial_guess
    list_x = [current_x]
    deriv_list = [derivative_func(current_x)]

    for _ in range(max_iter):
        previous_x = current_x
        gradient = derivative_func(previous_x)
        current_x = previous_x - multiplier * gradient

        list_x.append(current_x)
        deriv_list.append(derivative_func(current_x))

        if abs(current_x - previous_x) < precision:
            break

    return current_x, list_x, deriv_list

# Function to calculate cost values for an experiment
def calculate_cost_values(experiment):
    cost_vals = np.array([x**2 for x in experiment])
    return cost_vals[0] - cost_vals

# Run gradient descent
n = 100
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=n)
cost_reduction_low_gamma = calculate_cost_values(low_gamma[1])

# Plotting the reduction in cost for each iteration
plt.figure(figsize=[10, 6])
iterations = np.arange(1, n+2)  # Incrementing to match the length of gradient descent output

plt.plot(iterations, cost_reduction_low_gamma, label='Low Gamma: multiplier=0.0005', color='blue')

plt.title('Reduction in Cost for Each Iteration', fontsize=15)
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Cost Reduction', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()


```

```python
import numpy as np
from sympy import symbols, sin, sqrt, diff

# Define symbols for x and y
x, y = symbols('x y')

# Define the function f(x, y)
def f(x, y):
    return sin(sqrt(x**2 + y**2))

# Calculate partial derivatives with respect to x and y
def fpx(x_val, y_val):
    return diff(f(x, y), x).evalf(subs={x: x_val, y: y_val})

def fpy(x_val, y_val):
    return diff(f(x, y), y).evalf(subs={x: x_val, y: y_val})

# Given multipliers and initial parameters
multiplier = 0.1
params = np.array([2, 3])  # Example initial parameters

max_iter = 100  # Set the maximum number of iterations for optimization

# Optimization loop
for n in range(max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])

    # Check if gradients are valid before updating
    if gradient_x is not None and gradient_y is not None:
        gradients = np.array([gradient_x, gradient_y])
        params = params - multiplier * gradients
    else:
        # Handle the case where gradients are None (not computed properly)
        print("Gradients are None. Check the gradient calculation functions.")
        break

# Results or further processing
print("Optimized parameters:", params)
```

## Mean Squared Error: a cost function for regression problems

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
